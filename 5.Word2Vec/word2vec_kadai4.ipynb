# Word2Vec（事前学習済みモデル）

このノートブックでは、事前学習済みのWord2Vecモデルを読み込み、単語の類似度やベクトル演算（アナロジー）を試します。

```python
# 必要なライブラリ
import os
import gensim.downloader as api
from gensim.models import KeyedVectors

# モデル設定
# 1) Gensimの公開モデルを使う場合（ネットワーク接続が必要）
GENSIM_MODEL_NAME = "glove-wiki-gigaword-100"

# 2) ローカルの事前学習済みモデルを使う場合（.kv / .bin / .txt）
LOCAL_MODEL_PATH = ""  # 例: "./GoogleNews-vectors-negative300.bin"

# モデルの読み込み
if LOCAL_MODEL_PATH and os.path.exists(LOCAL_MODEL_PATH):
    if LOCAL_MODEL_PATH.endswith(".kv"):
        wv = KeyedVectors.load(LOCAL_MODEL_PATH, mmap="r")
    else:
        wv = KeyedVectors.load_word2vec_format(
            LOCAL_MODEL_PATH, binary=LOCAL_MODEL_PATH.endswith(".bin")
        )
    print("Loaded local model:", LOCAL_MODEL_PATH)
else:
    wv = api.load(GENSIM_MODEL_NAME)
    print("Loaded gensim model:", GENSIM_MODEL_NAME)

print("vocab size:", len(wv.key_to_index))
```

## 1. 類似単語の確認

代表的な単語に対して、類似単語を確認します。

```python
for word in ["king", "queen", "cat", "dog", "computer"]:
    if word in wv:
        print(word, ":", wv.most_similar(word, topn=5))
    else:
        print(word, "は語彙に存在しません")
```

## 2. ベクトル演算（アナロジー）

5通り以上の例で、単語ベクトルの演算結果を確認します。

```python
analogies = [
    ("king", "man", "woman"),
    ("paris", "france", "italy"),
    ("tokyo", "japan", "france"),
    ("car", "wheel", "bird"),
    ("big", "bigger", "small"),
    ("good", "better", "bad"),
]

for a, b, c in analogies:
    if a in wv and b in wv and c in wv:
        result = wv.most_similar(positive=[a, c], negative=[b], topn=5)
        print(f"{a} - {b} + {c} = {result[:3]}")
    else:
        print(f"{a}, {b}, {c} のいずれかが語彙にありません")
```

## 3. 追加の確認（類似度）

単語間の類似度を数値で確認します。

```python
pairs = [
    ("king", "queen"),
    ("cat", "dog"),
    ("car", "automobile"),
    ("computer", "keyboard"),
    ("coffee", "tea"),
]

for w1, w2 in pairs:
    if w1 in wv and w2 in wv:
        sim = wv.similarity(w1, w2)
        print(f"sim({w1}, {w2}) = {sim:.3f}")
    else:
        print(f"{w1} または {w2} が語彙にありません")
```
